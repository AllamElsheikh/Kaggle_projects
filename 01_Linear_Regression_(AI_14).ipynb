# My First Machine Learning Algorithm

**Linear Regression**

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Preparing Data
x = np.array([5, 15, 25, 35, 45, 55, 65]).reshape((-1,1))
y = np.array([15, 20,30, 45, 50, 40 ,80])
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
r_sq = model.score(x, y)
print(f"coefficient of determination: {r_sq*100} %")
```

```python
# Plot the data

# create a 8x6 (inches) figure, dpi means the resolution of the figure (dpi=80, 8
plt.figure(figsize=(8,5), dpi=60)
plt.style.use("Solarize_Light2")

# show the graph
plt.scatter(x, y )
```

```python
# create a 8x6 (inches) figure, dpi means the resolution of the figure (dpi=80, 8
plt.figure(figsize=(8,6), dpi=60)
plt.style.use("Solarize_Light2")

# show the graph
plt.scatter(x, y )
plt.plot(x, y )
```

```python
# Make Predictions

y_pred = model.predict(x)
y_pred
```

```python
# create a 8x6 (inches) figure, dpi means the resolution of the figure (dpi=80, 8
plt.figure(figsize=(8,6), dpi=60)
plt.style.use("Solarize_Light2")

# show the graph
plt.scatter(x, y )
plt.plot(x, y_pred )
```

```python
x_new = np.array([30, 50,60, 70, 80, 90 ,100]).reshape(-1,1)
```

```python
y_new = model.predict(x_new)
y_new
```

```python

# show the graph
plt.scatter(x, y )
plt.plot(x_new, y_new )
```

 ## Y= m*x + c           ----->>  model.predict(x)
 ## m = slope            ----->>  model.coef_
## c = Y axis inercept  ----->>  model.intercept_

 ## **Y = model.intercept_ + (model.coef_ * x)**
 ## x must be vertical so we used reshape(-1,1)
