# **Indian Diabetes Classification ANN**

This dataset describes the medical records for Pima Indians
and whether or not each patient will have an onset of diabetes within
ve years.

Fields description follow:

preg = Number of times pregnant

plas = Plasma glucose concentration a 2 hours in an oral glucose tolerance test

pres = Diastolic blood pressure (mm Hg)

skin = Triceps skin fold thickness (mm)

test = 2-Hour serum insulin (mu U/ml)

mass = Body mass index (weight in kg/(height in m)^2)

pedi = Diabetes pedigree function

age = Age (years)

class = Class variable (1:tested positive for diabetes, 0: tested negative for diabetes)

# first neural network with keras make predictions

```python
import numpy as np
import pandas as pd

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# load the dataset
cols=['Pregnancies',	'Glucose',	'BloodPressure'	,'SkinThickness',	'Insulin',	'BMI',	'DiabetesPedigreeFunction',	'Age'	,'Outcome']
dataset = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv',header=None, delimiter=',' ,names=cols)

dataset.head()
```

```python
dataset.shape
```

```python
X = dataset.drop('Outcome' ,axis=1)
X.head(2)
```

```python
X = dataset.iloc[:,0:-1]
y = dataset.iloc[:,-1]
y.head(2)
```

# define the keras model

```python
model = Sequential()

model.add(Dense(128, input_shape=(8,), activation='relu')) # hidden layer 1
model.add(Dense(256, activation='relu')) # hidden layer 2

model.add(Dense(1, activation='sigmoid')) # Output Layer


# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# fit the keras model on the dataset
model.fit(X, y, epochs=250, batch_size=10, verbose=1)
```

```python
predictions = model.predict(X)
# First 10 predections
for i in range(10):
	print('%2d  =>   %1.1f (expected %d)' % (i+1, predictions[i], y[i]))
```

```python
# make class predictions with the model
predictions = (model.predict(X) > 0.5).astype(int)
# First 10 predections
predictions[:10]
```
